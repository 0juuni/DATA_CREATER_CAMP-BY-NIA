```python
# https://velog.io/@brillantescene/CNN-owndata
from PIL import Image
import os
import natsort
import csv
import re

img_size = (128, 128)

def resize_img(img_path):
    img_lists = os.listdir(img_path)
    img_lists = natsort.natsorted(img_lists)
    for img_name in img_lists:
        print(f'{img_path}{img_name}')
        image = Image.open(f'{img_path}{img_name}')
        image = image.resize(img_size)
        image.save(f'{img_path}{img_name}')
```

```python
resize_img("/content/content/content/Data_4(테스트_트레인 분류 + 사진 자르기)/test/bottoms/")
resize_img("/content/content/content/Data_4(테스트_트레인 분류 + 사진 자르기)/test/cap_and_hat/")
resize_img("/content/content/content/Data_4(테스트_트레인 분류 + 사진 자르기)/test/outerwear/")
resize_img("/content/content/content/Data_4(테스트_트레인 분류 + 사진 자르기)/test/shoes/")
resize_img("/content/content/content/Data_4(테스트_트레인 분류 + 사진 자르기)/test/tops/")

resize_img("/content/content/content/Data_4(테스트_트레인 분류 + 사진 자르기)/train/bottoms/")
resize_img("/content/content/content/Data_4(테스트_트레인 분류 + 사진 자르기)/train/cap_and_hat/")
resize_img("/content/content/content/Data_4(테스트_트레인 분류 + 사진 자르기)/train/outerwear/")
resize_img("/content/content/content/Data_4(테스트_트레인 분류 + 사진 자르기)/train/shoes/")
resize_img("/content/content/content/Data_4(테스트_트레인 분류 + 사진 자르기)/train/tops/")
```
